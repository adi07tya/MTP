```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('russia.csv',names=['creation_date','tweet_id','retweet_count','like_count'])
#df = pd.read_csv('adgpi.csv',names=['creation_date','tweet_id','retweet_count','like_count'])
#df = pd.read_csv('us.csv',names=['creation_date','tweet_id','retweet_count','like_count'])
#df = pd.read_csv('australian.csv',names=['creation_date','tweet_id','retweet_count','like_count'])
#df = pd.read_csv('british.csv',names=['creation_date','tweet_id','retweet_count','like_count'])
#df = pd.read_csv('france.csv',names=['creation_date','tweet_id','retweet_count','like_count'])
```

```python
var = list(df['retweet_count'][df['tweet_id'] == 892661702059331584]) #russia
#var = list(df['retweet_count'][df['tweet_id'] == 893027198906408962]) #india
#var = list(df['retweet_count'][df['tweet_id'] == 892427278038310912]) #us
#var = list(df['retweet_count'][df['tweet_id'] == 892893953766195202]) #australia
#var = list(df['retweet_count'][df['tweet_id'] == 893428827136942082]) #britain
#var = list(df['retweet_count'][df['tweet_id'] == 892803706793721858]) #france
```

```python
val = [var[0]]
```

```python
val
```

```python
for i in range(0,len(var)-1):
    res = abs(var[i+1]-var[i])
    val.append(res)
```

```python
val
```

```python
y = list(range(0,len(val)))
```

```python
y
```

```python
plt.xlabel('Time_Interval')
plt.ylabel('Retweet_Count')
plt.title('RUSSIA')
plt.grid()
plt.plot(y,val)
```

```python
fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig('russia.png', dpi=100)
```
